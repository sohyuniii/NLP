```python
import json
import pandas as pd
import numpy as np

with open('review_NV.json') as data_file:    
    review = json.load(data_file)    
len(review)
```

```python
import gensim
from gensim import corpora, models
from gensim.models import CoherenceModel
import warnings
warnings.filterwarnings('ignore')
```

```python
stopWords = ['있','잇','하','되','좋']
```

```python
token = ['다리', '미','되', '구비', '센스', '좋']
no_stopword = [w for w in token if not w in stopWords]
no_stopword
```

```python
review[2]
```

```python
for rev in review : 
    sen_lst = []
    for sen in rev[1] :
        no_stopword = [w for w in sen if not w in stopWords]
        sen_lst.append(no_stopword)
        rev[1] = sen_lst
```

```python
review[2]
```

```python
texts = []
for rev in review:
    for sen in rev[1] :
        if len(sen)>1 :
            texts.append(sen)
        
print('문장 갯수 :',len(texts))
texts[:3]
```

```python
dic = corpora.Dictionary(texts)
```

```python
corpus = [dic.doc2bow(text) for text in texts]
```

```python
ldamodel6 = gensim.models.ldamodel.LdaModel(corpus, num_topics = 6, id2word = dic) # default
```

```python
ldamodel6.show_topics(num_words=10) #토픽을 구성하는 주요단어들
```

```python
ldamodel5 = gensim.models.ldamodel.LdaModel(corpus, num_topics = 5, id2word = dic) # default
```

```python
ldamodel5.show_topics(num_words=10) #토픽을 구성하는 주요단어들
```

```python
ldamodel4 = gensim.models.ldamodel.LdaModel(corpus, num_topics = 4, id2word = dic) # default
```

```python
ldamodel4.show_topics(num_words=10) #토픽을 구성하는 주요단어들
```
